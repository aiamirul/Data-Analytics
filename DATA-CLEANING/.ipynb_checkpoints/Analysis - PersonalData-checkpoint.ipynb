```python
# Load the Pandas libraries with alias 'pd' 
import pandas as pd 
import numpy as np
# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 
WorkSpaceName = 'AKPK-Personal'
data = pd.read_csv("personal.csv") 
# Preview the first 5 lines of the loaded data 
print(data.shape)
print(data.columns)
data.head(100)Tenure
```

```python
data = data[data['BAL'].isnull() == 0]
data = data[data['Tenure'] != 'X']
data['Tenure'] = data['Tenure'].astype(int)
data = data[data['Tenure'] > 0]

data['BAL']  = data['BAL'].str.strip()
data['BAL']  = data['BAL'].str.replace(',','')
data['BAL'] = data['BAL'].astype(float)

```

```python

data['IST']  = data['IST'].str.strip()
data['IST']  = data['IST'].str.replace(',','')
data['IST']  = data['IST'].str.replace('-','0')
data['IST'] = data['IST'].astype(float)
data = data[data['IST'] > 0]
```

```python
print(data.Gender.value_counts())

print(data.Gender.unique())
data['Gender']  = data['Gender'].str.replace('Male','1')
data['Gender']  = data['Gender'].str.replace('Female','0')
data['Gender'] = data['Gender'].astype(int)
data['Gender'].value_counts()
```

```python

data['profilingscore'] = data['profilingscore'].astype(int)
data['profilingscore'] = data[data['profilingscore'] > 0]
data['profilingscore'] = data['profilingscore'].dropna()
data
```

```python
print(data.profilingrating.value_counts())

data['profilingrating']  = data['profilingrating'].str.replace('A','1')
data['profilingrating']  = data['profilingrating'].str.replace('B','2')
data['profilingrating']  = data['profilingrating'].str.replace('C','3')
data['profilingrating']  = data['profilingrating'].str.replace('D','4')
data['profilingrating'] = data['profilingrating'].astype(int)


data['NumberOfDependent'] = data['NumberOfDependent'].astype(int)
```

# Save Correlation Matrix to PNG

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cols = data.columns.shape[0]
imgsize = cols/2.5 # SET sIZING BASED ON COLUMNS AMOUNT
fig, ax = plt.subplots(sharex=True, figsize=(imgsize,imgsize))

# .2f is the precision of Correlation
sns.heatmap(data.corr(method='pearson'), annot=True, fmt='.2f', 
            cmap=plt.get_cmap('coolwarm'), cbar=False, ax=ax)

ax.set_yticklabels(ax.get_yticklabels(), rotation="horizontal")
plt.savefig(WorkSpaceName+'_Correlation_Matrix.png', bbox_inches='tight', pad_inches=0.0)
```
